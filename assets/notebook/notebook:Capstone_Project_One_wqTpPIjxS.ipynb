```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'

source = requests.get(url) #get the html text from the site
soup = BeautifulSoup(source.content, 'html.parser') # parse it so you can read it

table = soup.find('table')

rows = table.find_all('tr')

a = []

for row in rows:  # getting the data out of the table by looping throw the rows and columns
    for cell in row.find_all('td'):
        cell.string = cell.string.split("\n")[0]
        if cell.text == "Not assigned" or cell.text == "": # set a empty or 'Not assigned' cell to zero
            cell.string = "0"
        a.append(cell.text)



a = np.array(a)
a = a.reshape(len(a)//3, 3) # respape the dataset and put it into an df
    
df_main = pd.DataFrame(a)

df_main = df_main.rename(columns={ 0 : "Postal Code", 1 : "Borough", 2 : "Neighborhood"})
df_main = df_main[df_main.Borough != "0"]
df_main = df_main[df_main.Neighborhood != "0"] # drop all rows with no relevant information

#df.to_csv(r'./output.csv', index = False)

print(df_main)
print(df_main.shape)
```
